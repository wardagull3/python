# splitting a string into multiple columns

```python
df=pd.DataFrame({'Name':['Warda Gull','Malaika Gull','Nimra Hayat'],
                 'Age':[20,19,21],
                 'Qualification':['BS','BS','BS'],
                 'CGPA':[3.5,3.7,3.6]})

df
```

```python
df[['1st name','Last name']]=df.Name.str.split(' ',expand=True)
df
```

```python
df=df[['Name','Age','Qualification','CGPA']]
df
```
