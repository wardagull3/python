```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
flower= sns.load_dataset('iris')
flower
```

```python
flower.isnull().sum()
```

```python
flower.columns
```

```python
sns.countplot(data=flower,x='species')
```

```python
sns.countplot(data=flower,x='species', hue='species')
```

```python
sns.scatterplot(data=flower,x='sepal_length',y='sepal_width', hue='species')
```

```python
sns.barplot(data=flower,x='sepal_length',y='sepal_width', hue='species')
```

```python
sns.boxplot(data=flower,x='sepal_length',y='sepal_width', hue='species')
```

```python
sns.stripplot(data=flower,x='sepal_length',y='sepal_width', hue='species')
```

```python
sns.violinplot(data=flower,x='sepal_length',y='sepal_width', hue='species')
```

```python
sns.boxenplot(data=flower,x='sepal_length',y='sepal_width', hue='species')
```

```python
sns.pairplot(flower)
```

```python
corr= flower.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("Heatmap")
```

```python
flower.hist()
```

```python
flower.hist(bins=10,figsize=(10,10), grid=False)
```

```python
plt.scatter(data=flower, x='sepal_length', y='sepal_width')
```

```python
flower.sepal_length.value_counts().plot(kind='bar')
```

```python
g= sns.FacetGrid(flower, 
                col='species', 
                hue='sepal_length',
                margin_titles=True,
                # palette={1:'seagreen', 0:'gray'}
                )
g=g.map(plt.scatter, 'sepal_length', 'species', edgecolor='w').add_legend()
```

```python
# How to change the titles of the x and y axis
sns.boxenplot(data=flower,x='sepal_length',y='sepal_width', hue='species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Boxen Plot')
```
